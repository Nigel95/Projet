```python
import pandas as pd
```

```python
dataset_median_age = pd.read_csv('d:/Formations/Data_Analysis/Data_Analyst_Fullstack/Projects/Demo_day/0_Source_Datasets/total_median_age.csv')
display(dataset_median_age)
```

```python
dataset_median_age_unpivot = pd.melt(
                                        dataset_median_age,
                                        id_vars = "economy",
                                        var_name = "Year",
                                        value_name="Median_age"    
                                                                )
display(dataset_median_age_unpivot)
```

```python
dataset_median_age_unpivot.rename(columns={"economy": "Country Code"}, inplace=True)

display(dataset_median_age_unpivot)
```

```python
dataset_median_age_unpivot["Year"] = dataset_median_age_unpivot["Year"].apply(lambda x : x[2:])
display(dataset_median_age_unpivot)
```

```python
dataset_median_age_unpivot.to_csv('Dataset_median_age.csv')
```
