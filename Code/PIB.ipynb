```python
df_Birth_Rate = pd.read_csv("total_birth_rate.csv", encoding="unicode_escape")
```

```python
df_Birth_Rate.head()
```

```python
df_Birth_Rate_unpivot = pd.melt(df_Birth_Rate, id_vars=['economy'], var_name='Year', value_name='Birth Rate')

# Affichez le résultat
df_Birth_Rate_unpivot.head()
```

```python
df_Birth_Rate_unpivot.rename(columns={'economy': 'Country Code'}, inplace=True)

# Affichez le résultat
df_Birth_Rate_unpivot.head()
```

```python
df_Birth_Rate_unpivot["Year"] = [yearacronym.replace("YR", "") for yearacronym in df_Birth_Rate_unpivot["Year"]]
display(df_Birth_Rate_unpivot) 
```

```python
df_PIB_tronque.head()
```

```python
df_PIB_tronque["Year"] = df_PIB_tronque["Year"].apply(lambda x: x[:4]) 
display(df_PIB_tronque)
```

```python
display(df_Birth_Rate_unpivot)
```

```python
df_Birth_Rate_PIB = pd.merge(df_Birth_Rate_unpivot, df_PIB_tronque, on=['Country Code', 'Year'])
display(df_Birth_Rate_PIB)
```

```python
colonnes = df_Birth_Rate_PIB.columns.tolist()

# Déplacez la colonne "Country name" de la position 5 à la position 1
colonnes.insert(4, colonnes.pop(6))

# Réorganisez les colonnes dans le DataFrame
df_Birth_Rate_PIB = df_Birth_Rate_PIB[colonnes]

# Affichez le résultat
display(df_Birth_Rate_PIB)
```
